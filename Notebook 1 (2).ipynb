# 1. Selección y descripción del conjunto de datos

# 1.1 Selección Conjunto de Datos

Para el desarrollo del proyecto se seleccionaron cinco bases de datos oficiales provenientes del ICFES. Se tomó la base más reciente del examen Saber Pro correspondiente al periodo 2024 y, adicionalmente, las cuatro bases del examen Saber 11 correspondientes a los años en que estos mismos estudiantes presentaron dicha prueba, es decir, 2018 y 2019 (cada año con sus dos periodos).

A partir de una base adicional proporcionada por el ICFES, denominada base de cruce Saber 11 – Saber Pro, fue posible identificar y vincular los registros correspondientes al mismo estudiante en ambos exámenes. Como resultado de este proceso de integración, se consolidó una base de datos final en la que cada registro corresponde a un estudiante e incluye tanto sus variables socioeconómicas como los resultados obtenidos en ambos exámenes.

La selección de este conjunto de datos responde al interés de abordar el análisis desde la perspectiva del Objetivo de Desarrollo Sostenible (ODS) 4: Educación de Calidad. En un país como Colombia, donde persisten importantes brechas educativas y donde la inversión en educación es limitada en comparación con otras economías, resulta fundamental analizar los factores que influyen no solo en la calidad de la educación básica y media, sino también en la continuidad de los estudiantes en la educación superior.

Este conjunto de datos permite evaluar múltiples dimensiones —sociales, económicas y académicas— y, gracias al cruce entre ambos exámenes estatales, ofrece la posibilidad de realizar un seguimiento longitudinal del estudiante. Esta característica es particularmente valiosa, ya que pocas bases de datos permiten observar la trayectoria educativa desde la educación media hasta la educación superior.

Lo anterior posibilita un análisis más integral del desempeño académico y facilita la construcción de modelos predictivos que no se limiten a explicar el resultado de un único examen, sino que permitan anticipar trayectorias educativas y potenciales resultados futuros. Esto puede aportar información relevante para que entidades estatales y privadas tomen decisiones más estratégicas sobre la asignación y focalización de recursos en el sector educativo, buscando generar un impacto real y sostenible.

Finalmente, se optó por utilizar bases de datos del ICFES debido a que son fuentes oficiales, con procesos previos de depuración y validación, lo que garantiza un nivel adecuado de confiabilidad, consistencia y respaldo institucional para el desarrollo de un proyecto de analítica serio.

# 1.2 Alineación con ODS de Colombia

El proyecto se encuentra directamente alineado con el Objetivo de Desarrollo Sostenible (ODS) 4: Garantizar una educación inclusiva, equitativa y de calidad y promover oportunidades de aprendizaje durante toda la vida para todos. En el contexto colombiano, este objetivo adquiere especial relevancia debido a las persistentes brechas estructurales en el sistema educativo.

En Colombia existen diferencias significativas en los resultados académicos entre estudiantes de instituciones educativas públicas y privadas, particularmente en la educación básica y media. Estas brechas suelen estar asociadas a factores como condiciones socioeconómicas, calidad de la infraestructura, disponibilidad de recursos pedagógicos, formación docente y acceso a apoyo académico complementario. Como consecuencia, el desempeño en pruebas estandarizadas como Saber 11 refleja desigualdades que posteriormente pueden trasladarse al acceso, permanencia y desempeño en la educación superior.

El análisis propuesto permite identificar qué variables socioeconómicas y académicas están más relacionadas con el desempeño en las pruebas nacionales y, especialmente, con la continuidad hacia la educación superior, medida a través del cruce entre Saber 11 y Saber Pro. Esto es fundamental para comprender si el sistema educativo está promoviendo realmente trayectorias de formación sostenidas o si, por el contrario, las brechas iniciales tienden a profundizarse con el tiempo.

Adicionalmente, Colombia enfrenta un reto estructural relacionado con la disponibilidad de mano de obra altamente calificada y verdaderamente especializada. Aunque ha habido avances en cobertura en educación superior, persisten desafíos en términos de calidad, pertinencia y culminación de los programas académicos. Un análisis basado en datos permite identificar patrones tempranos de riesgo académico, factores asociados al bajo desempeño y variables que influyen en la progresión educativa, lo cual puede contribuir al diseño de políticas públicas más focalizadas y estrategias institucionales de acompañamiento.

En este sentido, el proyecto no solo se alinea con el ODS 4 en términos de medición de calidad educativa, sino también en su contribución potencial a la reducción de desigualdades, la mejora de la permanencia estudiantil y el fortalecimiento del capital humano del país. # Modelos de machine learning

```python
import re

def _cols_like(df, patterns):
    """Devuelve columnas que coinciden con cualquiera de los patrones regex."""
    cols = []
    for c in df.columns:
        c_str = str(c)
        for p in patterns:
            if re.search(p, c_str, flags=re.IGNORECASE):
                cols.append(c)
                break
    return cols


def build_df_saber11(
    df,
    ignore_saber11=None,
    keep_targets=("nivel_saber11", "punt_global_saber11"),
    drop_percentiles=True,
    drop_desempenos=True,
    drop_puntajes_competencias=True,
):
    """
    Para modelar Saber11:
    - Elimina variables de SaberPro
    - Elimina desempeños, puntajes por competencia y percentiles de Saber11
    - Conserva los targets definidos en keep_targets si existen
    """
    ignore_saber11 = set(ignore_saber11 or [])
    df_out = df.copy()

    # Eliminar columna vacía si existe
    if "" in df_out.columns:
        df_out = df_out.drop(columns=[""], errors="ignore")

    # 1) Eliminar todo lo asociado a SaberPro
    drop_saberpro = set(_cols_like(df_out, [
        r"_saberpro\b",
        r"\bsaberpro\b",
        r"^mod_",
        r"^percentil_.*saberpro\b",
        r"^percentil_global_saberpro\b",
        r"^punt_global_saberpro\b",
        r"^nivel_saberpro\b",
        r"^percentil_nbc\b",
    ]))

    # 2) Eliminar variables de desempeño/puntajes/percentiles de Saber11
    drop_saber11_perf = set()

    if drop_desempenos:
        drop_saber11_perf |= set(_cols_like(df_out, [r"^desemp_"]))

    if drop_puntajes_competencias:
        drop_saber11_perf |= set(_cols_like(df_out, [
            r"^punt_",
            r"^recaf_punt_"
        ]))

    if drop_percentiles:
        drop_saber11_perf |= set(_cols_like(df_out, [r"^percentil_"]))

    # 3) Eliminar columnas explícitas indicadas por el usuario
    drop_explicit = {c for c in ignore_saber11 if c in df_out.columns}

    # 4) Mantener targets si existen
    keep_targets = [t for t in keep_targets if t in df_out.columns]

    drop_all = (drop_saberpro | drop_saber11_perf | drop_explicit) - set(keep_targets)

    df_out = df_out.drop(columns=sorted(drop_all), errors="ignore")
    return df_out


def build_df_saberpro(
    df,
    ignore_saberpro=None,
    keep_targets=("nivel_saberpro", "punt_global_saberpro"),
    drop_modulos=True,
    drop_percentiles_saberpro=True,
    drop_niveles_modulos=True,
    drop_puntajes_modulos=True,
    drop_global_generic=False,
):
    """
    Para modelar SaberPro:
    - Conserva variables de contexto, incluyendo Saber11
    - Elimina módulos, percentiles y puntajes/niveles modulares de SaberPro
    - Conserva los targets definidos en keep_targets si existen
    """
    ignore_saberpro = set(ignore_saberpro or [])
    df_out = df.copy()

    if "" in df_out.columns:
        df_out = df_out.drop(columns=[""], errors="ignore")

    drop_saberpro_perf = set()

    # 1) Eliminar módulos SaberPro
    if drop_modulos:
        drop_saberpro_perf |= set(_cols_like(df_out, [r"^mod_"]))

    # 2) Eliminar percentiles de SaberPro
    if drop_percentiles_saberpro:
        drop_saberpro_perf |= set(_cols_like(df_out, [
            r"^percentil_global_saberpro\b",
            r"^percentil_.*saberpro\b",
            r"^percentil_nbc\b",
            r"^percentil_global\b",
        ]))

    # 3) Eliminar niveles/puntajes modulares
    if drop_niveles_modulos:
        drop_saberpro_perf |= set(_cols_like(df_out, [
            r"_desem\b",
            r"_desem_"
        ]))

    if drop_puntajes_modulos:
        drop_saberpro_perf |= set(_cols_like(df_out, [
            r"_punt\b",
            r"_punt_"
        ]))

    # 4) Eliminar columnas explícitas
    drop_explicit = {c for c in ignore_saberpro if c in df_out.columns}

    # 5) Opcional: eliminar punt_global genérico
    if drop_global_generic:
        drop_saberpro_perf |= set(_cols_like(df_out, [r"^punt_global\b"]))

    # 6) Mantener targets
    keep_targets = [t for t in keep_targets if t in df_out.columns]

    drop_all = (drop_saberpro_perf | drop_explicit) - set(keep_targets)

    df_out = df_out.drop(columns=sorted(drop_all), errors="ignore")
    return df_out


# =========================
# USO SEGURO
# =========================

# Si no definiste listas manuales, se crean vacías
ignore_saber11 = []
ignore_saberpro = []

df_saber11 = build_df_saber11(
    df,
    ignore_saber11=ignore_saber11,
    keep_targets=("nivel_saber11", "punt_global_saber11"),
)

df_saberpro = build_df_saberpro(
    df,
    ignore_saberpro=ignore_saberpro,
    keep_targets=("nivel_saberpro", "punt_global_saberpro"),
    drop_percentiles_saberpro=True,
    drop_global_generic=False,
)

print("df original:", df.shape)
print("df_saber11:", df_saber11.shape)
print("df_saberpro:", df_saberpro.shape)
```

```python
import pandas as pd
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (
    RandomForestRegressor,
    GradientBoostingRegressor,
    AdaBoostRegressor
)
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# =========================================================
# 1) ORDINAL SPECS
# =========================================================
ORDINAL_SPECS = {
    "fami_estratovivienda_saber11": [
        "Sin Estrato", "Estrato 1", "Estrato 2", "Estrato 3",
        "Estrato 4", "Estrato 5", "Estrato 6"
    ],
    "fami_estratovivienda_saberpro": [
        "Sin Estrato", "Estrato 1", "Estrato 2", "Estrato 3",
        "Estrato 4", "Estrato 5", "Estrato 6"
    ],
    "estu_nse_establecimiento": ["NSE1", "NSE2", "NSE3", "NSE4"],
    "estu_nse_individual_saber11": ["NSE1", "NSE2", "NSE3", "NSE4"],
    "estu_nse_individual_saberpro": ["NSE1", "NSE2", "NSE3", "NSE4"],
    "fami_educacionmadre_saber11": [
        "Ninguno", "Primaria incompleta", "Primaria completa",
        "Secundaria (Bachillerato) incompleta",
        "Secundaria (Bachillerato) completa",
        "Técnica o tecnológica incompleta",
        "Técnica o tecnológica completa",
        "Educación profesional incompleta",
        "Educación profesional completa",
        "Postgrado", "No sabe", "No Aplica"
    ],
    "fami_educacionpadre_saber11": [
        "Ninguno", "Primaria incompleta", "Primaria completa",
        "Secundaria (Bachillerato) incompleta",
        "Secundaria (Bachillerato) completa",
        "Técnica o tecnológica incompleta",
        "Técnica o tecnológica completa",
        "Educación profesional incompleta",
        "Educación profesional completa",
        "Postgrado", "No sabe", "No Aplica"
    ],
    "fami_educacionmadre_saberpro": [
        "Ninguno", "Primaria incompleta", "Primaria completa",
        "Secundaria (Bachillerato) incompleta",
        "Secundaria (Bachillerato) completa",
        "Técnica o tecnológica incompleta",
        "Técnica o tecnológica completa",
        "Educación profesional incompleta",
        "Educación profesional completa",
        "Postgrado", "No sabe", "No Aplica"
    ],
    "fami_educacionpadre_saberpro": [
        "Ninguno", "Primaria incompleta", "Primaria completa",
        "Secundaria (Bachillerato) incompleta",
        "Secundaria (Bachillerato) completa",
        "Técnica o tecnológica incompleta",
        "Técnica o tecnológica completa",
        "Educación profesional incompleta",
        "Educación profesional completa",
        "Postgrado", "No sabe", "No Aplica"
    ],
    "estu_dedicacionlecturadiaria": [
        "No leo por entretenimiento",
        "30 minutos o menos",
        "Entre 30 y 60 minutos",
        "Entre 1 y 2 horas",
        "Más de 2 horas"
    ],
    "estu_dedicacioninternet": [
        "No Navega Internet",
        "30 minutos o menos",
        "Entre 30 y 60 minutos",
        "Entre 1 y 3 horas",
        "Más de 3 horas"
    ],
    "fami_comecarnepescadohuevo": [
        "Nunca o rara vez comemos eso",
        "1 o 2 veces por semana",
        "3 a 5 veces por semana",
        "Todos o casi todos los días"
    ],
    "fami_comecerealfrutoslegumbre": [
        "Nunca o rara vez comemos eso",
        "1 o 2 veces por semana",
        "3 a 5 veces por semana",
        "Todos o casi todos los días"
    ],
    "fami_comelechederivados": [
        "Nunca o rara vez comemos eso",
        "1 o 2 veces por semana",
        "3 a 5 veces por semana",
        "Todos o casi todos los días"
    ],
    "fami_personashogar": ["1 a 2", "3 a 4", "5 a 6", "7 a 8", "9 o más"],
    "fami_cuartoshogar": ["Uno", "Dos", "Tres", "Cuatro", "Cinco", "Seis o mas"],
    "fami_numlibros": ["0 A 10 LIBROS", "11 A 25 LIBROS", "26 A 100 LIBROS", "MÁS DE 100 LIBROS"],
    "fami_situacioneconomica": ["Peor", "Igual", "Mejor"],
    "desemp_ingles": ["A-", "A1", "A2", "B+", "B1"],
    "nivel_saber11": ["Deficiente", "Bajo", "Regular", "Alto", "Excelente"],
    "nivel_saberpro": ["Deficiente", "Bajo", "Regular", "Alto", "Excelente"],
    "estu_horassemanatrabaja_saberpro": [
        "0", "Menos de 10 horas", "Entre 11 y 20 horas",
        "Entre 21 y 30 horas", "Más de 30 horas"
    ],
    "estu_semestrecursa": [
        "01", "02", "03", "04", "05", "06",
        "07", "08", "09", "10", "11", "12 o más"
    ],
    "inst_caracter_academico": [
        "TÉCNICA PROFESIONAL",
        "INSTITUCIÓN TECNOLÓGICA",
        "INSTITUCIÓN UNIVERSITARIA",
        "UNIVERSIDAD"
    ]
}


# =========================================================
# 2) FUNCIONES AUXILIARES
# =========================================================
def mape_seguro(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    mask = y_true != 0
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


def crear_preprocesador(X, ordinal_specs):
    ordinal_cols = [col for col in ordinal_specs.keys() if col in X.columns]
    ordinal_categories = [ordinal_specs[col] for col in ordinal_cols]

    cat_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
    cat_cols = [col for col in cat_cols if col not in ordinal_cols]

    num_cols = X.select_dtypes(include=[np.number]).columns.tolist()

    preprocessor = ColumnTransformer(
        transformers=[
            (
                "ordinal",
                OrdinalEncoder(
                    categories=ordinal_categories,
                    handle_unknown="use_encoded_value",
                    unknown_value=-1
                ),
                ordinal_cols
            ),
            (
                "categorical",
                OrdinalEncoder(
                    handle_unknown="use_encoded_value",
                    unknown_value=-1
                ),
                cat_cols
            ),
            ("numerical", "passthrough", num_cols)
        ],
        remainder="drop"
    )

    return preprocessor


def obtener_modelos_regresion(random_state=42):
    return {
        "LinearRegression": LinearRegression(),
        "Ridge": Ridge(),
        "Lasso": Lasso(),
        "ElasticNet": ElasticNet(),
        "DecisionTreeRegressor": DecisionTreeRegressor(random_state=random_state),
        "RandomForestRegressor": RandomForestRegressor(random_state=random_state),
        "GradientBoostingRegressor": GradientBoostingRegressor(random_state=random_state),
        "KNeighborsRegressor": KNeighborsRegressor(),
        "SVR": SVR(),
        "AdaBoostRegressor": AdaBoostRegressor(random_state=random_state)
    }


def obtener_param_grids():
    return {
        "LinearRegression": {},
        "Ridge": {
            "model__alpha": [0.01, 0.1, 1.0, 10.0, 100.0]
        },
        "Lasso": {
            "model__alpha": [0.001, 0.01, 0.1, 1.0, 10.0]
        },
        "ElasticNet": {
            "model__alpha": [0.001, 0.01, 0.1, 1.0],
            "model__l1_ratio": [0.2, 0.5, 0.8]
        },
        "DecisionTreeRegressor": {
            "model__max_depth": [None, 5, 10, 20],
            "model__min_samples_split": [2, 5, 10],
            "model__min_samples_leaf": [1, 2, 4]
        },
        "RandomForestRegressor": {
            "model__n_estimators": [100, 200],
            "model__max_depth": [None, 10, 20],
            "model__min_samples_split": [2, 5],
            "model__min_samples_leaf": [1, 2]
        },
        "GradientBoostingRegressor": {
            "model__n_estimators": [100, 200],
            "model__learning_rate": [0.01, 0.05, 0.1],
            "model__max_depth": [3, 5],
            "model__subsample": [0.8, 1.0]
        },
        "KNeighborsRegressor": {
            "model__n_neighbors": [3, 5, 7, 9],
            "model__weights": ["uniform", "distance"],
            "model__p": [1, 2]
        },
        "SVR": {
            "model__C": [0.1, 1, 10],
            "model__epsilon": [0.01, 0.1, 0.5],
            "model__kernel": ["rbf", "linear"]
        },
        "AdaBoostRegressor": {
            "model__n_estimators": [50, 100, 200],
            "model__learning_rate": [0.01, 0.1, 1.0],
            "model__loss": ["linear", "square"]
        }
    }


# =========================================================
# 3) ESTA ES LA FUNCIÓN QUE TE FALTABA
# =========================================================
def evaluar_modelos_predictivos(df, target_col, ordinal_specs, test_size=0.2, random_state=42):
    data = df.copy()

    X = data.drop(columns=[target_col])
    y = data[target_col]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )

    preprocessor = crear_preprocesador(X_train, ordinal_specs)
    modelos = obtener_modelos_regresion(random_state=random_state)

    resultados = []

    for nombre, modelo in modelos.items():
        pipeline = Pipeline([
            ("preprocessor", preprocessor),
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler()),
            ("model", modelo)
        ])

        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict(X_test)

        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test, y_pred)
        mape = mape_seguro(y_test, y_pred)

        resultados.append({
            "modelo": nombre,
            "MAE": mae,
            "MSE": mse,
            "RMSE": rmse,
            "R2": r2,
            "MAPE (%)": mape
        })

    resultados_df = pd.DataFrame(resultados).sort_values(by="MAPE (%)", ascending=True).reset_index(drop=True)
    return resultados_df


# =========================================================
# 4) GRIDSEARCH SOBRE LOS 5 MEJORES
# =========================================================
def gridsearch_top_5_modelos(
    df,
    target_col,
    ordinal_specs,
    test_size=0.2,
    random_state=42,
    cv=3,
    scoring="neg_mean_absolute_error"
):
    data = df.copy()

    X = data.drop(columns=[target_col])
    y = data[target_col]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )

    resultados_base = evaluar_modelos_predictivos(
        df=data,
        target_col=target_col,
        ordinal_specs=ordinal_specs,
        test_size=test_size,
        random_state=random_state
    )

    top_5 = resultados_base.head(5)["modelo"].tolist()

    modelos = obtener_modelos_regresion(random_state=random_state)
    param_grids = obtener_param_grids()
    preprocessor = crear_preprocesador(X_train, ordinal_specs)

    resumen = []
    mejores_modelos = {}

    for nombre in top_5:
        modelo = modelos[nombre]

        pipeline = Pipeline([
            ("preprocessor", preprocessor),
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler()),
            ("model", modelo)
        ])

        param_grid = param_grids[nombre]

        if not param_grid:
            pipeline.fit(X_train, y_train)
            mejor_modelo = pipeline
            y_pred = mejor_modelo.predict(X_test)

            mae = mean_absolute_error(y_test, y_pred)
            mse = mean_squared_error(y_test, y_pred)
            rmse = np.sqrt(mse)
            r2 = r2_score(y_test, y_pred)
            mape = mape_seguro(y_test, y_pred)

            mejores_modelos[nombre] = {
                "best_estimator": mejor_modelo,
                "best_params": {},
                "best_cv_score": np.nan
            }

            resumen.append({
                "modelo": nombre,
                "best_params": {},
                "CV Score": np.nan,
                "MAE": mae,
                "MSE": mse,
                "RMSE": rmse,
                "R2": r2,
                "MAPE (%)": mape
            })
            continue

        grid = GridSearchCV(
            estimator=pipeline,
            param_grid=param_grid,
            cv=cv,
            scoring=scoring,
            n_jobs=-1
        )

        grid.fit(X_train, y_train)

        mejor_modelo = grid.best_estimator_
        y_pred = mejor_modelo.predict(X_test)

        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test, y_pred)
        mape = mape_seguro(y_test, y_pred)

        mejores_modelos[nombre] = {
            "best_estimator": mejor_modelo,
            "best_params": grid.best_params_,
            "best_cv_score": grid.best_score_
        }

        resumen.append({
            "modelo": nombre,
            "best_params": grid.best_params_,
            "CV Score": grid.best_score_,
            "MAE": mae,
            "MSE": mse,
            "RMSE": rmse,
            "R2": r2,
            "MAPE (%)": mape
        })

    resumen_df = pd.DataFrame(resumen).sort_values(by="MAPE (%)", ascending=True).reset_index(drop=True)

    return resultados_base, top_5, mejores_modelos, resumen_df
```
